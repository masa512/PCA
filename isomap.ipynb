```python
import numpy as np
from sklearn.datasets import make_swiss_roll, load_digits
import matplotlib.pyplot as plt
from sklearn import manifold
```

# Part 1 : ISOMAP (using Geodesic distance + MDS)

```python
# We will load 3D swissroll dataset N = 100 representative of the smooth manifold

X,t= make_swiss_roll(n_samples = 2000, noise = 0.5) # We'll 

print(X.shape,": The size of swissroll")

# Visualize the datapoints
f = plt.figure(figsize=(10,10))
ax = f.add_subplot(projection = '3d')
ax.scatter(X[:,0],X[:,1],X[:,2],c=t, cmap = 'jet')


```

```python
# We will use the built-in function for ISOMAP
M = manifold.Isomap(n_neighbors=6, n_components=2)
embedding = M.fit_transform(X)
```

```python
# Plot the result in the same scatter condition
print(embedding.shape)
f = plt.figure(figsize = (10,10))
plt.scatter(embedding[:,0],embedding[:,1],c=t, cmap = 'jet')
plt.xlabel('Dim 1')
plt.ylabel('Dim 2')
```

# Part 2: Using Digit dataset to cluster using digits dataset 

```python
data = load_digits()
X = data.data
label = data.target

print('Data dimension:',X.shape)


# We will try to make some visualizations of what we have
fig, axes = plt.subplots(nrows = 10, ncols = 10,figsize = (10,10))

for i, ax in enumerate(axes.flatten()):
  ax.axis("off") # Turns the border and stuff off
  #Since the image is vectorized to length 64, we will reshape
  ax.imshow(X[i,:].reshape((8,8)),cmap=plt.cm.binary)

fig.suptitle('Visualization of digits dataset')
```

```python
m = manifold.LocallyLinearEmbedding(n_neighbors=6,n_components=2)
Xtrans = m.fit_transform(X[:,:])
```

```python
plt.scatter(Xtrans[:,0],Xtrans[:,1],c=label)
```
